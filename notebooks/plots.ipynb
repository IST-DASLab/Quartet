```python
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt  # Added for plotting

str_size_to_int = {
    "30M": 30482560,
    "50M": 49556736,
    "100M": 102777856,
    "200M": 203188480,
    "430M": 443068288,
    "800M": 822151168,
}

# Yellow, Blue, Red, Green
main_colors = ["#FFA700", "#0057E7", "#D62D20", "#008744"]

# Create a figure displaying four squares, each with a different main color.
fig, ax = plt.subplots(figsize=(6, 2))
ax.set_xlim(0, 4)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis("off")

# Draw one square per color
for i, color in enumerate(main_colors):
    square = plt.Rectangle((i, 0), 1, 1, facecolor=color)
    ax.add_patch(square)

fig.savefig("../figures/color_scheme.pdf", transparent=True, bbox_inches="tight")
plt.show()
```

```python
from scipy.optimize import least_squares

FORCE_GAMMA_1 = False
FORCE_BETA_1 = False

def bf16_model(base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / N ** alpha + np.exp(b) / D ** beta) ** gamma + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [0, 5, 10]
beta_grid = [0.0, 0.5, 1.0]
e_grid = [0.0, 0.5] # , 1.0, 1.5]
gamma_grid = [0.0, 0.5, 1.0]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for beta0 in beta_grid:
                for e0 in e_grid:
                    for gamma0 in gamma_grid:
                    
                        initial_guess = [a0, alpha0, b0, beta0, e0, gamma0]
                        
                        result = least_squares(
                            fun=residuals,
                            x0=initial_guess,
                            args=(bf16_data,),
                            loss='huber',
                            f_scale=1e-4,  # This is the delta parameter for Huber loss
                        )
                        
                        if result.cost < best_cost:
                            best_cost = result.cost
                            best_result = result

a_fit, alpha_fit, b_fit, beta_fit, e_fit, gamma_fit = best_result.x
best_base_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit)}")
print(f"alpha   = {alpha_fit}")
print(f"B       = {np.exp(b_fit)}")
print(f"beta    = {beta_fit}")
print(f"E       = {np.exp(e_fit)}")
print(f"gamma   = {gamma_fit}")
```

```python
def backward_model(backward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effd, = backward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N) ** alpha + np.exp(b) / (D * effd) ** beta) ** gamma + np.exp(e)


backward_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    }
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "AlbertTsengQuantizer"
]


def residuals_backward(backward_params, base_params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll apply a robust loss (Huber) to these via least_squares(..., loss='huber').
    """
    residuals_list = []
    for point in data:
        # Avoid numerical issues: ensure model(...) > 0 before log
        L_model = backward_model(backward_params, base_params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
            
        residuals_list.append(r)
    return np.array(residuals_list)

effd_grid = [1.0]

best_effb_result = None
best_cost = np.inf

for effd0 in tqdm(effd_grid, desc="effd0"):
    initial_guess = [effd0]
    
    result = least_squares(
        fun=lambda backward_params: residuals_backward(backward_params, best_base_params, backward_data),
        x0=initial_guess,
        loss='huber',
        f_scale=1e-4,  # Delta parameter for Huber loss
    )
    
    if result.cost < best_cost:
        best_cost = result.cost
        best_result = result

effd_fit, = best_result.x
best_backward_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"Effd    = {effd_fit}")
```

```python
def forward_model(forward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effn, = forward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N * effn) ** alpha + np.exp(b) / D ** beta) ** gamma + np.exp(e)

def forward_backward_model(forward_params, backward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effn, = forward_params
    effd, = backward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N * effn) ** alpha + np.exp(b) / (D * effd) ** beta) ** gamma + np.exp(e)

forward_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    }
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "QuEST" and bwd == "BF16"
]


def residuals(forward_params, base_params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    """
    residuals_list = []
    for point in data:
        # Avoid numerical issues: ensure model(...) > 0 before log
        L_model = forward_model(forward_params, base_params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

effn_grid = [0.5]

best_result = None
best_cost = np.inf

for eff0 in tqdm(effn_grid, desc="effn0"):
    initial_guess = [eff0]
    
    result = least_squares(
        fun=lambda forward_params: residuals(forward_params, best_base_params, forward_data),
        x0=initial_guess,
        loss='huber',
        f_scale=1e-4,
    )
    
    if result.cost < best_cost:
        best_cost = result.cost
        best_result = result

effn_fit, = best_result.x
best_forward_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"Effn    = {effn_fit}")
```

```python
def unified_model(ns, ds, forward, backward):
    if forward and backward:
        return forward_backward_model(best_forward_params, best_backward_params, best_base_params, ns, ds)
    elif forward:
        return forward_model(best_forward_params, best_base_params, ns, ds)
    elif backward:
        return backward_model(best_backward_params, best_base_params, ns, ds)
    else:
        return bf16_model(best_base_params, ns, ds)
```

```python
def model_size_to_speedups(n: np.ndarray, batch_size: int = 32) -> np.ndarray:
    """
    Interpolates speedups for an array of model sizes n.
    Returns an array of shape (len(n), 2) with interpolated speedups.
    """
    speedups = all_speedups[batch_size]
    
    n_flat = n.flatten()
    
    x = np.array(list(speedups.keys()))
    y = np.array([speedups[k] for k in x])  # shape (len(x), 2)

    # Interpolate for each column (speedup type)
    out_forward = np.interp(n_flat, x, y[:, 0], left=0, right=y[-1, 0])
    out_backward = np.interp(n_flat, x, y[:, 1], left=0, right=y[-1, 1])
    
    return out_forward.reshape(n.shape), out_backward.reshape(n.shape)
```

### Speedups model


| Operation | Speedup relative to baseline (FP8) | FP4:FP8 | FP8:FP4 | FP4:FP4 |
|-|-|-|-|-|
| Forward Pass and Inference | $\text{spd}_{forward}(P_{forward}, N)$ | 2.0 | 1.0 | 2.0 |
| Backward Pass | $\text{spd}_{backward}(P_{backward}, N)$ | 1.0 | 2.0 | 2.0 |
| Training | $(\frac{1}{3}\cdot \text{spd}_{forward}(P_{forward}, N)^{-1} + \frac{2}{3}\cdot \text{spd}_{backward}(P_{backward}, N)^{-1})^{-1}$ | 1.2 | 1.5 | 2.0 |

### Stochastic backward

| Quantization scheme | Inference constraint | Training Constraint | Predicted Loss | Linear Model |
|-|-|-|-|-|
| FP8:FP8 | $N = N_{max}$ | $N \cdot D = N_{max}\cdot D_{max}$ | $L(N_{max}, D_{max}, FP8, FP8)$ | $L(N_{max}, D_{max}, FP8, FP8)$ |
| FP8:FP4 | $N = N_{max}$ | $N/sptr \cdot D = N_{max}\cdot D_{max}$ | $L(N_{max}, D_{max}\cdot sptr, FP8, FP4)$| $L(N_{max}, D_{max}\cdot 1.5, FP8, FP4)$
| FP4:FP8 | $\frac{N}{spfw} = N_{max}$ | $N/sptr \cdot D = N_{max}\cdot D_{max}$ | $L(N_{max}\cdot spfw, D_{max} \cdot sptr / spfw, FP4, FP8)$ | $L(N_{max}\cdot 2, D_{max} \cdot 0.6, FP4, FP8)$ |

```python
def depth_from_name(name):
    idx = int(name.split(".")[3])
    depth = 5 - idx
    if "attn" in name:
        depth += 0.5
    return depth

# Define quantizers to plot
quantizers_to_plot = ["AlbertTsengQuantizer", "AlbertTsengDeterministic" , "EdenQuantizer"]
colors = [main_colors[3], main_colors[2], main_colors[1]]
labels = ["SR AbsMax", "RTN AbsMax", "RTN AbsMax EDEN"]

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))

# Process data for each quantizer
all_x_sorted = []
for i, quantizer in enumerate(quantizers_to_plot):
    data = alignments[quantizer]
    
    # Initialize lists to store processed data
    x_values = []
    cosine_means = []
    cosine_stds = []
    misalignment_means = []
    misalignment_stds = []
    
    # Axis 1: cosine similarity
    # Process each key
    for key, values in data.items():
        x_val = depth_from_name(key)
        x_values.append(x_val)
        cosine_means.append(values['cosine_mean'])
        cosine_stds.append(values['cosine_std'])
        misalignment_means.append(1 - values['alignment_mean'])
        misalignment_stds.append(values['alignment_std'])
    
    # Sort the data based on x_values
    sorted_indices = np.argsort(x_values)
    x_sorted = np.array(x_values)[sorted_indices]
    all_x_sorted.append(x_sorted)
    cosine_means_sorted = np.array(cosine_means)[sorted_indices]
    cosine_stds_sorted = np.array(cosine_stds)[sorted_indices]
    misalignment_means_sorted = np.array(misalignment_means)[sorted_indices]
    misalignment_stds_sorted = np.array(misalignment_stds)[sorted_indices]
    
    # Plot the data for this quantizer
    ax1.plot(x_sorted, cosine_means_sorted, marker='o' if quantizer != "EdenQuantizer" else 'x', linestyle='-', color=colors[i], label=labels[i])
    ax2.plot(x_sorted, misalignment_means_sorted, marker='o' if quantizer != "EdenQuantizer" else 'x', linestyle='-', color=colors[i], label=labels[i])
    
    # Add shaded region for standard deviation
    ax1.fill_between(
        x_sorted,
        cosine_means_sorted - cosine_stds_sorted,
        cosine_means_sorted + cosine_stds_sorted,
        color=colors[i],
        alpha=0.2
    )
    ax2.fill_between(
        x_sorted,
        misalignment_means_sorted - misalignment_stds_sorted,
        misalignment_means_sorted + misalignment_stds_sorted,
        color=colors[i],
        alpha=0.2
    )

# Gather all unique integer x values for ticks, skip non-integers
all_x_ticks = set()
for x_sorted in all_x_sorted:
    for x in x_sorted:
        if np.isclose(x, round(x)):
            all_x_ticks.add(int(round(x)))
x_ticks_sorted_unique = sorted(all_x_ticks)

# Customize the plot
ax1.axhline(y=1.0, color='black', linestyle='--')
ax1.set_xlabel('Backprop Distance, Transformer Blocks')
ax1.set_xlim(0, 6)
ax1.set_ylabel('Cosine Similarity')
ax1.set_xticks(x_ticks_sorted_unique)  # only integer ticks
ax1.legend(fontsize=12)

ax2.axhline(y=0.0, color='black', linestyle='--')
ax2.set_xlabel('Backprop Distance, Transformer Blocks')
ax2.set_xlim(0, 6)
ax2.set_ylabel(r'Misalignment')
ax2.set_xticks(x_ticks_sorted_unique)  # only integer ticks
# ax2.legend()


# Define line styles for different model sizes
linestyles = {
    "30M": "-",
    "50M": "--",
    "100M": "-."
}

# Plot for all model sizes on first subplot (Stochastic and Deterministic)
for size, linestyle in linestyles.items():
    # Normalize factor based on model size
    if size == "30M":
        norm_factor = 30482560
    elif size == "50M":
        norm_factor = 49556736
    elif size == "100M":
        norm_factor = 99113472
    
    # Stochastic Backward - BF16
    ax3.plot(
        list(map(lambda x: x * 512 * 512 / norm_factor, size_fwd_bwd_iters_loss[size]["BF16"]["AlbertTsengQuantizer"].keys())),
        [s - b for s, b in zip(size_fwd_bwd_iters_loss[size]["BF16"]["AlbertTsengQuantizer"].values(), size_fwd_bwd_iters_loss[size]["BF16"]["BF16"].values())],
        marker="o", color=colors[0], linestyle=linestyle, 
        label=None
    )
    
    # Deterministic Backward - BF16
    ax3.plot(
        list(map(lambda x: x * 512 * 512 / norm_factor, size_fwd_bwd_iters_loss[size]["BF16"]["AlbertTsengDeterministic"].keys())),
        [d - b for d, b in zip(size_fwd_bwd_iters_loss[size]["BF16"]["AlbertTsengDeterministic"].values(), size_fwd_bwd_iters_loss[size]["BF16"]["BF16"].values())],
        marker="o", color=colors[1], linestyle=linestyle,
        label="_nolegend_"
    )
    
    # Aligned Backward - BF16
    if size == "30M" and "AlignedAlbertTsengQuantizer" in quantizers_to_plot:
        ax3.plot(
            [2861*512*512/norm_factor, 5722*512*512/norm_factor, 11444*512*512/norm_factor, 22888*512*512/norm_factor, 45776*512*512/norm_factor, 91552*512*512/norm_factor],
            [3.42335 - 3.41571, 3.30172 - 3.29368, 3.21491 - 3.20785, 3.15403 - 3.147, 3.10919 - 3.09895, 3.07244 - 3.0609],
            marker="^", color="purple", linestyle=linestyle,
            label="_nolegend_"
        )
        
    # Eden Backward - BF16
    if "EdenQuantizer" in quantizers_to_plot:
        ax3.plot(
            list(map(lambda x: x * 512 * 512 / norm_factor, size_fwd_bwd_iters_loss[size]["BF16"]["Eden128"].keys())),
            [d - b for d, b in zip(size_fwd_bwd_iters_loss[size]["BF16"]["Eden128"].values(), size_fwd_bwd_iters_loss[size]["BF16"]["BF16"].values())],
            marker="x", color=colors[2], linestyle=linestyle,
            label="_nolegend_"
        )


# Add scheme entries to legend for first subplot
ax3.plot([], [], color=colors[0], linestyle="-", marker="o", label="SR AbsMax")
ax3.plot([], [], color=colors[1], linestyle="-", marker="o", label="RTN AbsMax")
if "AlignedAlbertTsengQuantizer" in quantizers_to_plot:
    ax3.plot([], [], color="purple", linestyle="-", marker="o", label="RTN AbsMax PMA")
if "EdenQuantizer" in quantizers_to_plot:
    ax3.plot([], [], color=colors[2], linestyle="-", marker="x", label="RTN AbsMax EDEN")

# # Add model size entries to legend for first subplot
# for size, linestyle in linestyles.items():
#     ax3.plot([], [], color="black", linestyle=linestyle, label=f"{size}")

# Configure first subplot
ax3.set_xlim(20, 1000)
ax3.set_ylim(0, None)
ax3.set_xscale("log")
ax3.set_xlabel(r"D/N")
ax3.set_ylabel("C4 Val Loss Gap")
ax3.legend(loc="lower right")

methods_legend = ax3.legend(loc="lower left")
ax3.legend(loc="lower right", fontsize=12, handles=[
    ax3.plot([], [], color="black", linestyle=linestyles["30M"], label="N=30M")[0],
    ax3.plot([], [], color="black", linestyle=linestyles["50M"], label="N=50M")[0],
    ax3.plot([], [], color="black", linestyle=linestyles["100M"], label="N=100M")[0],
])
# ax3.add_artist(methods_legend)


labels = ['(a)', '(b)', '(c)']
for ax, label in zip((ax1, ax2, ax3), labels):
    ax.annotate(label, 
                xy=(0.5, -0.20),     # x=0.5 (center), y=-0.25 (below the x-axis)
                xycoords='axes fraction',
                ha='center', 
                va='top', 
                fontsize=16)

fig.tight_layout()
fig.savefig("../figures/cosine_and_alignment.pdf", bbox_inches="tight", transparent=True)


# Display the plot
plt.show()
```

```python
def bf16_hoffman_model(base_params, N, D):
    a, alpha, b, beta, e = base_params
    return np.exp(a) / N ** alpha + np.exp(b) / D ** beta + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals_hoffman(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_hoffman_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [0, 5, 10]
beta_grid = [0.0, 0.5, 1.0]
e_grid = [0.0, 0.5] # , 1.0, 1.5]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for beta0 in beta_grid:
                for e0 in e_grid:
                    initial_guess = [a0, alpha0, b0, beta0, e0]
                    
                    result = least_squares(
                        fun=residuals_hoffman,
                        x0=initial_guess,
                        args=(bf16_data,),
                        loss='huber',
                        f_scale=1e-4,  # This is the delta parameter for Huber loss
                    )
                    
                    if result.cost < best_cost:
                        best_cost = result.cost
                        best_result = result

a_fit_hoffman, alpha_fit_hoffman, b_fit_hoffman, beta_fit_hoffman, e_fit_hoffman = best_result.x
best_hoffman_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit_hoffman)}")
print(f"alpha   = {alpha_fit_hoffman}")
print(f"B       = {np.exp(b_fit_hoffman)}")
print(f"beta    = {beta_fit_hoffman}")
print(f"E       = {np.exp(e_fit_hoffman)}")
```

```python
def bf16_kaplan_model(base_params, N, D):
    a, alpha, b, e, gamma = base_params
    return (np.exp(a) / N ** alpha + np.exp(b) / D) ** gamma + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals_kaplan(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_kaplan_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [10, 15, 20]
e_grid = [0.0, 0.5]
gamma_grid = [0.5, 1.0]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for e0 in e_grid:
                for gamma0 in gamma_grid:
                    initial_guess = [a0, alpha0, b0, e0, gamma0]
                    
                    result = least_squares(
                        fun=residuals_kaplan,
                        x0=initial_guess,
                        args=(bf16_data,),
                        loss='huber',
                        f_scale=1e-4,  # This is the delta parameter for Huber loss
                    )
                    
                    if result.cost < best_cost:
                        best_cost = result.cost
                        best_result = result

a_fit_kaplan, alpha_fit_kaplan, b_fit_kaplan, e_fit_kaplan, gamma_fit_kaplan = best_result.x
best_kaplan_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit_kaplan)}")
print(f"alpha   = {alpha_fit_kaplan}")
print(f"B       = {np.exp(b_fit_kaplan)}")
print(f"E       = {np.exp(e_fit_kaplan)}")
print(f"gamma   = {gamma_fit_kaplan}")
```
