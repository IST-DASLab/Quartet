```python
import numpy as np
from tqdm.auto import tqdm
import matplotlib.pyplot as plt  # Added for plotting

str_size_to_int = {
    "30M": 30482560,
    "50M": 49556736,
    "100M": 102777856,
    "200M": 203188480,
    "430M": 443068288,
    "800M": 822151168,
}

# Yellow, Blue, Red, Green
main_colors = ["#FFA700", "#0057E7", "#D62D20", "#008744"]

# Create a figure displaying four squares, each with a different main color.
fig, ax = plt.subplots(figsize=(6, 2))
ax.set_xlim(0, 4)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis("off")

# Draw one square per color
for i, color in enumerate(main_colors):
    square = plt.Rectangle((i, 0), 1, 1, facecolor=color)
    ax.add_patch(square)

fig.savefig("../figures/color_scheme.pdf", transparent=True, bbox_inches="tight")
plt.show()
```

```python
from scipy.optimize import least_squares

FORCE_GAMMA_1 = False
FORCE_BETA_1 = False

def bf16_model(base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / N ** alpha + np.exp(b) / D ** beta) ** gamma + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [0, 5, 10]
beta_grid = [0.0, 0.5, 1.0]
e_grid = [0.0, 0.5] # , 1.0, 1.5]
gamma_grid = [0.0, 0.5, 1.0]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for beta0 in beta_grid:
                for e0 in e_grid:
                    for gamma0 in gamma_grid:
                    
                        initial_guess = [a0, alpha0, b0, beta0, e0, gamma0]
                        
                        result = least_squares(
                            fun=residuals,
                            x0=initial_guess,
                            args=(bf16_data,),
                            loss='huber',
                            f_scale=1e-4,  # This is the delta parameter for Huber loss
                        )
                        
                        if result.cost < best_cost:
                            best_cost = result.cost
                            best_result = result

a_fit, alpha_fit, b_fit, beta_fit, e_fit, gamma_fit = best_result.x
best_base_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit)}")
print(f"alpha   = {alpha_fit}")
print(f"B       = {np.exp(b_fit)}")
print(f"beta    = {beta_fit}")
print(f"E       = {np.exp(e_fit)}")
print(f"gamma   = {gamma_fit}")
```

```python
def backward_model(backward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effd, = backward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N) ** alpha + np.exp(b) / (D * effd) ** beta) ** gamma + np.exp(e)


backward_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    }
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "AlbertTsengQuantizer"
]


def residuals_backward(backward_params, base_params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll apply a robust loss (Huber) to these via least_squares(..., loss='huber').
    """
    residuals_list = []
    for point in data:
        # Avoid numerical issues: ensure model(...) > 0 before log
        L_model = backward_model(backward_params, base_params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
            
        residuals_list.append(r)
    return np.array(residuals_list)

effd_grid = [1.0]

best_effb_result = None
best_cost = np.inf

for effd0 in tqdm(effd_grid, desc="effd0"):
    initial_guess = [effd0]
    
    result = least_squares(
        fun=lambda backward_params: residuals_backward(backward_params, best_base_params, backward_data),
        x0=initial_guess,
        loss='huber',
        f_scale=1e-4,  # Delta parameter for Huber loss
    )
    
    if result.cost < best_cost:
        best_cost = result.cost
        best_result = result

effd_fit, = best_result.x
best_backward_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"Effd    = {effd_fit}")
```

```python
def forward_model(forward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effn, = forward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N * effn) ** alpha + np.exp(b) / D ** beta) ** gamma + np.exp(e)

def forward_backward_model(forward_params, backward_params, base_params, N, D):
    a, alpha, b, beta, e, gamma = base_params
    effn, = forward_params
    effd, = backward_params
    
    if FORCE_GAMMA_1:
        gamma = 1
    if FORCE_BETA_1:
        beta = 1
    return (np.exp(a) / (N * effn) ** alpha + np.exp(b) / (D * effd) ** beta) ** gamma + np.exp(e)

forward_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    }
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "QuEST" and bwd == "BF16"
]


def residuals(forward_params, base_params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    """
    residuals_list = []
    for point in data:
        # Avoid numerical issues: ensure model(...) > 0 before log
        L_model = forward_model(forward_params, base_params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

effn_grid = [0.5]

best_result = None
best_cost = np.inf

for eff0 in tqdm(effn_grid, desc="effn0"):
    initial_guess = [eff0]
    
    result = least_squares(
        fun=lambda forward_params: residuals(forward_params, best_base_params, forward_data),
        x0=initial_guess,
        loss='huber',
        f_scale=1e-4,
    )
    
    if result.cost < best_cost:
        best_cost = result.cost
        best_result = result

effn_fit, = best_result.x
best_forward_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"Effn    = {effn_fit}")
```

```python
def unified_model(ns, ds, forward, backward):
    if forward and backward:
        return forward_backward_model(best_forward_params, best_backward_params, best_base_params, ns, ds)
    elif forward:
        return forward_model(best_forward_params, best_base_params, ns, ds)
    elif backward:
        return backward_model(best_backward_params, best_base_params, ns, ds)
    else:
        return bf16_model(best_base_params, ns, ds)
```

| Operation | Speedup relative to baseline (FP8) | FP4:FP8 | FP8:FP4 | FP4:FP4 |
|-|-|-|-|-|
| Forward Pass and Inference | $\text{spd}_{forward}(P_{forward}, N)$ | 2.0 | 1.0 | 2.0 |
| Backward Pass | $\text{spd}_{backward}(P_{backward}, N)$ | 1.0 | 2.0 | 2.0 |
| Training | $(\frac{1}{3}\cdot \text{spd}_{forward}(P_{forward}, N)^{-1} + \frac{2}{3}\cdot \text{spd}_{backward}(P_{backward}, N)^{-1})^{-1}$ | 1.2 | 1.5 | 2.0 |

```python
def bf16_hoffman_model(base_params, N, D):
    a, alpha, b, beta, e = base_params
    return np.exp(a) / N ** alpha + np.exp(b) / D ** beta + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals_hoffman(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_hoffman_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [0, 5, 10]
beta_grid = [0.0, 0.5, 1.0]
e_grid = [0.0, 0.5] # , 1.0, 1.5]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for beta0 in beta_grid:
                for e0 in e_grid:
                    initial_guess = [a0, alpha0, b0, beta0, e0]
                    
                    result = least_squares(
                        fun=residuals_hoffman,
                        x0=initial_guess,
                        args=(bf16_data,),
                        loss='huber',
                        f_scale=1e-4,  # This is the delta parameter for Huber loss
                    )
                    
                    if result.cost < best_cost:
                        best_cost = result.cost
                        best_result = result

a_fit_hoffman, alpha_fit_hoffman, b_fit_hoffman, beta_fit_hoffman, e_fit_hoffman = best_result.x
best_hoffman_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit_hoffman)}")
print(f"alpha   = {alpha_fit_hoffman}")
print(f"B       = {np.exp(b_fit_hoffman)}")
print(f"beta    = {beta_fit_hoffman}")
print(f"E       = {np.exp(e_fit_hoffman)}")
```

```python
def bf16_kaplan_model(base_params, N, D):
    a, alpha, b, e, gamma = base_params
    return (np.exp(a) / N ** alpha + np.exp(b) / D) ** gamma + np.exp(e)


bf16_data = [
    {
        "N": str_size_to_int[N],
        "D": steps * 512 * 512,
        "L": loss
    } 
    for N, iso_size in size_fwd_bwd_iters_loss.items()
    for fwd, iso_fwd in iso_size.items()
    for bwd, iso_bwd in iso_fwd.items()
    for steps, loss in iso_bwd.items()
        if fwd == "BF16" and bwd == "BF16"
]


def residuals_kaplan(params, data):
    """
    Residuals in log-space: log(L_i) - log(model_prediction).
    We'll use the delta parameter for the Huber loss in least_squares.
    """
    residuals_list = []
    for point in data:
        L_model = bf16_kaplan_model(params, point['N'], point['D'])
        if L_model <= 0:
            # If the model predicts a non-positive value, penalize it heavily
            r = 1000
        else:
            r = (np.log(point['L']) - np.log(L_model))
        residuals_list.append(r)
    return np.array(residuals_list)

a_grid = [0, 5, 10]
alpha_grid = [0.0, 0.5, 1.0]
b_grid = [10, 15, 20]
e_grid = [0.0, 0.5]
gamma_grid = [0.5, 1.0]

best_result = None
best_cost = np.inf


for a0 in tqdm(a_grid, desc="a0"):
    for alpha0 in tqdm(alpha_grid, leave=False, desc="alpha0"):
        for b0 in tqdm(b_grid, leave=False, desc="b0"):
            for e0 in e_grid:
                for gamma0 in gamma_grid:
                    initial_guess = [a0, alpha0, b0, e0, gamma0]
                    
                    result = least_squares(
                        fun=residuals_kaplan,
                        x0=initial_guess,
                        args=(bf16_data,),
                        loss='huber',
                        f_scale=1e-4,  # This is the delta parameter for Huber loss
                    )
                    
                    if result.cost < best_cost:
                        best_cost = result.cost
                        best_result = result

a_fit_kaplan, alpha_fit_kaplan, b_fit_kaplan, e_fit_kaplan, gamma_fit_kaplan = best_result.x
best_kaplan_params = best_result.x
    
print(f"Best cost: {best_cost}")
print(f"Best result: {best_result.x}")

print(f"A       = {np.exp(a_fit_kaplan)}")
print(f"alpha   = {alpha_fit_kaplan}")
print(f"B       = {np.exp(b_fit_kaplan)}")
print(f"E       = {np.exp(e_fit_kaplan)}")
print(f"gamma   = {gamma_fit_kaplan}")
```
